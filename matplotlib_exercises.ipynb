```python
#Combine the figures you created in the last step into one large figure with 4 subplots.

plt.figure(figsize = (8, 8))

plt.subplot(2,2,1)
x = list(range(0, 8))
y = [n**(1/2) for n in x]
plt.scatter(x,y, s=30, c='red')
plt.grid(True, ls = '--')
plt.title('$y=x$')

plt.subplot(2,2,2)
x2 = list(range(-2, 3))
y2 = [n**3 for n in x2]
plt.scatter(x2,y2, s=30, c='blue')
plt.grid(True, ls = '--')
plt.title('$y=x^3$')

plt.subplot(2,2,3)
x3 = list(range(-5, 6))
y3 = [2**n for n in x3]
plt.scatter(x3,y3, s=30, c='green')
plt.grid(True, ls = '--')
plt.title('$y=2^x$')

plt.subplot(2,2,4)
x4 = list(range(-10, -1)) + list(range(0,11))
y4 = [1/(n+1) for n in x4]
plt.scatter(x4,y4, s=30, c='orange')
plt.grid(True, ls = '--')
plt.title('$y=1/(x+1)$')

plt.show
```

```python
#Combine the figures you created in the last step into one figure where each of the 4 equations has a
#different color for the points. Be sure to include a legend and an appropriate title for the figure.

plt.figure(figsize = (8, 8))

x = list(range(0, 8))
y = [n**(1/2) for n in x]
plt.scatter(x,y, s=30, c='red', label='$y=x$')

x2 = list(range(-2, 3))
y2 = [n**3 for n in x2]
plt.scatter(x2,y2, s=30, c='blue', label='$y=x^3$')

x3 = list(range(-5, 6))
y3 = [2**n for n in x3]
plt.scatter(x3,y3, s=30, c='green', label='$y=2^x$')

x4 = list(range(-10, -1)) + list(range(0,11))
y4 = [1/(n+1) for n in x4]
plt.scatter(x4,y4, s=30, c='orange', label='$y=1/(x+1)$')

plt.grid(True, ls = '--')
plt.title('One Large Figure')
plt.legend()

plt.show
```
