```python
# Load the swiss dataset and read it's documentation. Create visualizations to answer the following questions:

swiss = data('swiss')
#data('swiss', show_doc = True)
```

```python
# Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
# (Choose a cutoff point for what constitutes catholic)
swiss['is_catholic'] = swiss['Catholic'] > 50
```

```python
# Does whether or not a province is Catholic influence fertility?
sns.relplot(data = swiss, x = 'Fertility', y = 'Catholic', hue = 'is_catholic')
```

```python
# What measure correlates most strongly with fertility?
# Catholic, Agriculture, Infant Mortality
fig, ax = plt.subplots(3,2, figsize = (10,8))

sns.scatterplot(data = swiss, x = 'Fertility', y = 'Agriculture', ax = ax[0,0])
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Examination', ax = ax[0,1])
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Education', ax = ax[1,0])
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Catholic', ax = ax[1,1])
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Infant.Mortality', ax = ax[2,0])
```

```python
#Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items 
# and the revenue produced by each.

chipotle = get_db_url(user, host, password, 'chipotle')
orders_df = pd.read_sql('SELECT * FROM orders', chipotle)
orders_df.item_price = orders_df.item_price.str.replace('$', '').replace().astype('float')
top_four = orders_df.groupby('item_name').item_price.sum().nlargest(4)
```

```python
top_four
```

```python
item_names = ['Chicken Bowl', 'Chicken Burrito', 'Steak Burrito', 'Steak Bowl']
item_revenues = [7342.73, 5575.82, 3851.43, 2260.19]

top_four_df = pd.DataFrame({'item_name': item_names,
                   'revenue': item_revenues})
```

```python
sns.barplot(data = top_four_df, x = 'item_name', y = 'revenue')
```
